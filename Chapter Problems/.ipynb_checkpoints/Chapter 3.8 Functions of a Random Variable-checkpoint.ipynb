#### 1. General Problem: Find PDF of Transformed R.V
$f_X(x) = 3x^{2}$ for $0<x<1$

$Y = g(x)$

$Y = 1 - X^{2}$

```python
#PDF of X
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(0,1,0.1)
pdf = 3*np.power(x,2)
plt.plot(x,pdf)
plt.title("PDF of X")
plt.ylabel("Probability")
plt.xlabel("X")
plt.show()
```

```python
#Find the support of Y
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(0,1,0.1)
y = 1-np.power(x,2)
plt.plot(x,y)
plt.title('$Y = 1 - X^{2}$')
plt.ylabel("Y")
plt.xlabel("X")
plt.show()
```

We can see g(x) is one-to-one over the support of X. The support of Y is $0<y<1$.

$X = g^{-1}(y) = \sqrt{1-Y}$ ($g(x)$ is invertible over $0<x<1$)

$f_Y(y) = f_x(g^{-1}(y)) * |\frac{dg^{-1}(y)}{dy}|$ (Transformation method)

$\frac{dg^{-1}(y)}{dy} \sqrt{1-Y} = -\frac{1}{2}(1-y)^{-\frac{1}{2}}$

$f_Y(y) =  3(\sqrt{1-y})^{2}*|-\frac{1}{2}(y - 1)^{-\frac{1}{2}}|$

$f_Y(y) =  \frac{3}{2}(1-y)^{\frac{1}{2}}$

$f_Y(y) =  \frac{3}{2}\sqrt{(1-y)}$ for $0<y<1$

```python
#PDF of Y
import numpy as np
import matplotlib.pyplot as plt
y = np.arange(0,1,0.1)
y = 1 - y
pdf = (3/2)*np.sqrt(y)
plt.plot(y,pdf)
plt.title("PDF of Y")
plt.ylabel("Probability")
plt.xlabel("Y")
plt.show()
```
